```python
import pandas as pd

test_df = pd.read_csv("../data/split/test.csv")
val_df = pd.read_csv("../data/split/val.csv")
# train_df = pd.read_csv("../data/split/train.csv", index_col=0)
```

```python
def check_label_dist(df, name):
    print(f"\n{name}")
    print(df["Label"].value_counts(normalize=True))

check_label_dist(val_df, "VAL")
check_label_dist(test_df, "TEST")
```

```python
cols = ["LineId"]

overlap = pd.merge(
    val_df[cols],
    test_df[cols],
    how="inner"
)

print("Overlapping rows:", len(overlap))
```
