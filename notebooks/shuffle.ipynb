```python
import pandas as pd
```

```python
normal_log = pd.read_csv('../result/openstack_normal.csv')
anomaly_log = pd.read_csv('../result/openstack_abnormal.log_structured.csv')
```

```python
normal_log['Label'] = 0
anomaly_log['Label'] = 1

data = pd.concat([normal_log, anomaly_log], ignore_index=True)
data = data.sample(frac=1, random_state=42).reset_index(drop=True)  # Shuffle the data
data.to_csv('../result/shuffled_openstack_logs.csv', index=False)
```
