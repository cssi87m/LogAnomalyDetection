```python
import pandas as pd
```

```python

# Merge logs
df1 = pd.read_csv(
    '../result/openstack_normal1.log_structured.csv',
    sep=',',
    quotechar='"',
    escapechar='\\',
    engine='python'
)
df2 = pd.read_csv(
    '../result/openstack_normal2.log_structured.csv',
    sep=',',
    quotechar='"',
    escapechar='\\',
    engine='python'
)

combined = pd.concat([df1, df2], ignore_index=True)
combined.to_csv('../result/openstack_normal.csv', index=False)
```
